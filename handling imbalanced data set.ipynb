#imbalanced dataaset with an unequal class distribution

a data set with unequal class distribution

```python
#importing libraries
import pandas as pd
import numpy as np
```

## Loading the dataset to pandas dataframe


```python
credit_card_section=pd.read_csv('/content/credit_data.csv')
credit_card_section.head()
```

```python
#typical classification problem
```

```python
#distribution of two classes
```

```python
credit_card_section['Class'].value_counts()
```

0-->legit
1-->fraud

```python
#separating the legit and fraudulent trensactions
```

```python
legit=credit_card_section[credit_card_section.Class==0]
fraud=credit_card_section[credit_card_section.Class==1]
# here we mention the class value according to legit and fruad
```

```python
legit.shape
```

```python
fraud.shape
```

#undersampling method

build a sample dataset containing similar distribution of legot and fraudulent transactions

number of fraud transaction-->239

```python
legit_sample=legit.sample(n=239)
legit_sample.shape
```

# concatenate two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset['Class'].value_counts()
```
