# ***MY FIRST MACHINE LEARNING MODEL***


# **LOAD DATA**

```python
import pandas as pd

df =pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

# DATA PREPARATION

## dataseparation as x and y ##

```python
y=df['logS']
y
```

```python
x=df.drop('logS',axis=1)
```

```python
x
```

## data splitting for x and y


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size=0.2 , random_state=100)
```

```python
x_train
```

```python
x_test
```

# **MODEL** **BUILIDING**

## **linear regression**

## training the model for linear regression


```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train , y_train)
```

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
print(y_lr_train_pred)
```

```python
y_lr_test_pred
```

```python
y_lr_train_pred
```

## Evaluate model performance

```python
from sklearn.metrics import mean_squared_error , r2_score
lr_train_mse = mean_squared_error(y_train , y_lr_train_pred)
lr_train_r2 = r2_score(y_train , y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test , y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)

print('LR MSE (train):' , lr_train_mse)
print('LR R2  (train):' , lr_train_r2 )
print('LR MSE (test) :' , lr_test_mse )
print('LR R2  (test) :' , lr_test_r2  )
```

```python
lr_result = pd.DataFrame(['LinearRegressionegression' , lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_result.columns=['Method','Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_result
```

# Random forest

## Training the model

randomforest uses for both regressor and as wellas classifier

and now we are building regressionmodel because the y which means a log  is a quantitative model if its a categorical model we should build a classifier model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train , y_train)
```

## apply the model for prediction

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)

```

## Evaluate model performance

```python
from sklearn.metrics import mean_squared_error , r2_score

rf_train_mse = mean_squared_error (y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train , y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 =r2_score(y_test , y_rf_test_pred)

print('RF Mse (train):' , rf_train_mse)
print('RF R2  (train):' , rf_train_r2 )
print('RF MSE(TEST):',  rf_test_mse)
print('RF R2 (TEST):' , rf_test_r2)
```

```python
rf_result=pd.DataFrame(['RandomForest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_result.columns=['Method' , 'Training MSE' , 'Training R2' , 'Test MSE' , 'Test R2' ]
rf_result
```
