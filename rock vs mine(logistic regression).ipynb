# importing the dependencies

```python
import numpy as np                                        # for mathematical operations
import pandas as pd                                       # for data processing
from sklearn.model_selection import train_test_split      # used to split the data into training and testing data
from sklearn.linear_model import LogisticRegression       # for logistic regression
from sklearn.metrics import accuracy_score                # for accuracy score

```

# Data collection and preprocessing

```python
#loading the dataset to a pandas dataframe

sonar_data = pd.read_csv('/content/sonar.csv', header=None) #if file doesnt have header you should mention header=none

sonar_data.head()       #shows the first five
```

##Number of rows and columns

```python
sonar_data.shape
```

```python
sonar_data.describe()   #describe--> statistical measure of data
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

## separating lables and data

```python
#In this step we are storing every data except 60th column in x variable thats why we use (dropcolumn60)

x= sonar_data.drop(columns=60, axis=1) #if you are dropping a row mention axis=0

# and stored that 60th column in y varaible
y=sonar_data[60]
```

```python
print(x)
print(y)
```

now you can clearly see first it prints x as 59 data and y gives 60th column as lable

## Training and test data spliting

```python
#stratify ensures that the proportion of each class in the original dataset is preserved in both the training and testing sets
# random state used to split in a particular way basically to repoduce the code as it is
# test size mentions the spliting size
x_train, x_test ,y_train, y_test = train_test_split(x,y, test_size=0.1, stratify=y, random_state=1)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Logistic regression

## model training

```python
model= LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

# model evalutaion

```python
#accuracy on the training data
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print(training_data_accuracy)
```

```python
#accuracy on the test data
```

```python
x_test_prediction=model.predict(x_test)
training_data_accuracy = accuracy_score(x_test_prediction,y_test)
print(training_data_accuracy)
```
